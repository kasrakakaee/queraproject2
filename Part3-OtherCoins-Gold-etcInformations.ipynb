```python
def LabelDf(x):
    df = x.copy()
    df['price_day_before'] = df['Close'].shift(1)
    df['Label'] = df.apply(lambda row: 1 if row['Close'] > row['price_day_before'] else 0 , axis=1)
    df.drop(columns=['price_day_before'])
    return df
```
